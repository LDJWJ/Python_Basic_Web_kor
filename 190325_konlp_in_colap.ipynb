```python
import requests as req
from urllib.request import urlopen
from bs4 import BeautifulSoup
import lxml
```

```python
url = 'https://movie.daum.net/moviedb/grade?movieId=101393&type=netizen&page=2'
page = req.get(url)
html = page.content
soup = BeautifulSoup(html, 'lxml')
soup.prettify()
```

```python
s1 = soup.find('ul', class_='list_review list_netizen')
s1
```

```python
s2 = s1.find_all('li')
s2
```

```python
Nreviews = []

for item in s2:
  s3 = item.find('p', class_='desc_review').text
  s3 = s3.replace(' \n\t\t\t\t\t\t\t\t\t\t', '')
  s3 = s3.replace('\n\t\t\t\t\t\t\t\t\t', '')
  s3 = s3.replace(' \n', '')
  s3 = s3.replace('\r', ' ')
  Nreviews.append(s3)
  
Nreviews
```

## 페이지 수 늘이기

```python
url = 'https://movie.daum.net/moviedb/grade?movieId=101393&type=netizen&page='
urllist = []

for i in range(1,270):
  urls = url+str(i)
  urllist.append(urls)
  
urllist
```

```python
import time
pasttime=time.time()
time.sleep(5)
print(runningtime(pasttime))
```

```python
def runningtime(pasttime):
  nowtime=time.time()
  runtime = nowtime - pasttime
  return(runtime)
```

```python
Nreviews = []

pasttime = time.time()


for i in urllist:
  time.sleep(5)
  print(i)
  page = req.get(i)
  html = page.content
  soup = BeautifulSoup(html, 'lxml')
  s1 = soup.find('ul', class_='list_review list_netizen')
  s2 = s1.find_all('li')
  for j in s2:
    s3 = j.find('p', class_='desc_review').text
    s3 = s3.replace('\t', '')
    s3 = s3.replace('\n', '')
    s3 = s3.replace(' \n\t\t\t\t\t\t\t\t\t\t', '')
    s3 = s3.replace(' \n\t\t\t\t\t\t\t\t\t', '')
    s3 = s3.replace('\r', '')
    s3 = s3.replace(' \n', '')
    Nreviews.append(s3)
  
  print(runningtime(pasttime))
  print('댓글 추가 완료')
  
print(Nreviews)
```

```python
import pandas as pd
```

```python
Netizen_Review = pd.DataFrame({"dat":Nreviews})
Netizen_Review.to_csv('Netizen_Review.csv', index=False)
```
